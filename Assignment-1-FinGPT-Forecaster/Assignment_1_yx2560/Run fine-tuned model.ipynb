```python
import finnhub
import yfinance as yf
from datetime import datetime, timedelta
import os
import pandas as pd
import json

def verify_finnhub_data(symbol):
    """
    Verify Finnhub data for a given symbol
    """
    client = finnhub.Client(api_key=os.environ["FINNHUB_KEY"])

    # Get company profile
    profile = client.company_profile2(symbol=symbol)
    print("\nCompany Profile:")
    print(json.dumps(profile, indent=2))

    # Get basic financials
    financials = client.company_basic_financials(symbol, 'all')
    print("\nBasic Financials:")
    if 'metric' in financials:
        print(json.dumps(financials['metric'], indent=2))

    # Get recent news
    end_date = datetime.now()
    start_date = end_date - timedelta(weeks=3)
    news = client.company_news(
        symbol,
        _from=start_date.strftime('%Y-%m-%d'),
        to=end_date.strftime('%Y-%m-%d')
    )
    print(f"\nRecent News (last 3 items):")
    for n in news[:3]:
        print(f"Date: {datetime.fromtimestamp(n['datetime']).strftime('%Y-%m-%d')}")
        print(f"Headline: {n['headline']}")
        print(f"Summary: {n['summary']}\n")

def verify_yfinance_data(symbol):
    """
    Verify yfinance data for a given symbol
    """
    # Get stock data for last 3 weeks
    end_date = datetime.now()
    start_date = end_date - timedelta(weeks=3)

    stock_data = yf.download(
        symbol,
        start=start_date.strftime('%Y-%m-%d'),
        end=end_date.strftime('%Y-%m-%d'),
        progress=False
    )

    print("\nStock Price Data:")
    print(stock_data['Close'].describe())

    # Get additional info
    ticker = yf.Ticker(symbol)
    print("\nFinancial Info:")
    info = ticker.info
    metrics = [
        'grossMargins',
        'returnOnAssets',
        'returnOnEquity',
        'totalDebt',
        'totalEquity'
    ]
    for metric in metrics:
        if metric in info:
            print(f"{metric}: {info[metric]}")

# Usage example:
if __name__ == "__main__":
    symbol = "AAPL"
    print(f"Verifying data for {symbol}")
    print("=" * 50)

    print("\nChecking Finnhub Data:")
    verify_finnhub_data(symbol)

    print("\nChecking YFinance Data:")
    verify_yfinance_data(symbol)
```

```python
# Check just Finnhub data
verify_finnhub_data("AAPL")

# Check just YFinance data
verify_yfinance_data("AAPL")
```
